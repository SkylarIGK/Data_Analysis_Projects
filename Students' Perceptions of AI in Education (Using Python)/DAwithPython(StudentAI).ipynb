```python
# Convert GPA to numeric and classify into GPA categories
df["GPA"] = pd.to_numeric(df["GPA"], errors="coerce")
df["GPA_Category"] = df["GPA"].apply(lambda x: "High GPA" if 8.5 <= x <= 10 else ("Low GPA" if 5.0 <= x <= 6.4 else None))

# Filter and calculate average AI usefulness
df_filtered = df.dropna(subset=["GPA_Category"])
avg_usefulness = df_filtered.groupby("GPA_Category")["Utility_grade"].mean()

# Plot
plt.figure(figsize=(5, 4))
bars = plt.bar(avg_usefulness.index, avg_usefulness, color=["#1f77b4", "#ff7f0e"], edgecolor="black")
plt.title("AI Usefulness Perception by GPA")
plt.xlabel("GPA Category")
plt.ylabel("Average AI Usefulness (1-10)")

# Add value labels
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1, f"{bar.get_height():.2f}", ha='center')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ylim(0, 10)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Convert GPA to numeric
df["GPA"] = pd.to_numeric(df["GPA"], errors="coerce")

# Categorize GPA into groups
def categorize_gpa(gpa):
    if 5.0 <= gpa <= 6.4:
        return "Low GPA"
    elif 6.5 <= gpa <= 8.4:
        return "Mid GPA"
    elif 8.5 <= gpa <= 10:
        return "High GPA"
    else:
        return None

df["GPA_Group"] = df["GPA"].apply(categorize_gpa)

# Drop rows with undefined GPA group or missing job replacement concern
df_filtered = df.dropna(subset=["GPA_Group", "Job_replacement"])

# Group by GPA group and calculate average concern level
avg_concern = df_filtered.groupby("GPA_Group")["Job_replacement"].mean().reindex(["Low GPA", "Mid GPA", "High GPA"])

# Plotting
plt.figure(figsize=(6, 5))
bars = plt.bar(avg_concern.index, avg_concern, color=['#1f77b4', '#ff7f0e', '#2ca02c'], edgecolor="black")

# Add value labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.05, f"{height:.2f}", ha='center', fontsize=10)

# Labels & Title
plt.title("Concern About Job Replacement by GPA Group", fontsize=14)
plt.xlabel("GPA Group")
plt.ylabel("Average Concern Level (1-5)")
plt.ylim(0, 5)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
